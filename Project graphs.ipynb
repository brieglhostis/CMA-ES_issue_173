### Idea

After finding the best feasible:

$$
x^* = argmin_{x \in \mathbb{R^d}} f(x)
$$
With the constrainst:
$$
g_i(x^*) \leq 0$$ $$ h_j(x^*) = 0
$$
for $i=0,...,n_g$ and $j=0,...,n_h$

We try and find the best feasible:

$$
x^{**} = argmin_{x \in \mathbb{R^d}} \sum_{g_i(x^*) > 0} g_i(x)^2
$$
With the constrainst:
$$
g_i(x^{**}) \leq 0$$ $$ h_j(x^{**}) = 0
$$
for $i=0,...,n_g$ and $j=0,...,n_h$

And then, we keep the best solution between $x^*$ and $x^{**}$.


```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal
```

```python

rg = 1.8
x = np.arange(-rg, rg, 0.001)
g1 = lambda x: x**2

X = np.arange(-1, 1, 0.01) * rg
Y = np.arange(-1, 1, 0.01) * rg
```

```python

c = np.array([0, 0])
f = lambda x: np.sum(np.square(x - c))
g = lambda x: [x[0]**2 - x[1], 0.5**0.5 - x[0]]

x_best_feasible = (0.78044918, 0.85427122)
x_post_opt = (0.70711628, 0.50001362)

Z = np.array([[f([x, y]) for x in X] for y in Y])

plt.figure(figsize=(10, 6))
plt.title("Optimization landscape (f(x))")
plt.scatter(c[0], c[1], s=200, marker='x', label="Objective function optimum")
plt.scatter(x_best_feasible[0], x_best_feasible[1], s=200, marker='x', label="Best feasible, f=1.34")
plt.scatter(x_post_opt[0], x_post_opt[1], s=200, marker='x', label="Post optimization, f=0.75")
plt.contour(X, Y, Z, levels=30, linestyles="dashed")
plt.plot(x, g1(x), label="g1=0")
plt.plot(2 * [np.sqrt(0.5)], [rg, -rg], label="g2=0")
plt.ylim((-0.5, rg))
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()

lam = np.array([0.267, 1.643])
Z = np.array([[f(np.array([x, y])) + np.sum(lam * np.array(g([x, y]))) for x in X] for y in Y])

plt.figure(figsize=(10, 6))
plt.title("Optimization landscape during last Lagrangian optimization (f(x) + lam * g(x))")
plt.scatter(x_best_feasible[0], x_best_feasible[1], s=200, marker='x', label="Best feasible, f=1.34")
plt.scatter(x_post_opt[0], x_post_opt[1], s=200, marker='x', label="Post optimization, f=0.75")
plt.contour(X, Y, Z, levels=100, linestyles="dashed")
plt.plot(x, g1(x), label="g1=0")
plt.plot(2 * [np.sqrt(0.5)], [rg, -rg], label="g2=0")
plt.ylim((-0.5, rg))
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()

mean = np.array([0.70710413, 0.49999078])
cov = np.array([[8.54110940e-13, 8.93733321e-13], [8.93733321e-13, 6.15713576e-12]])

Z = np.array([[multivariate_normal.pdf(np.array([x, y]), mean=mean, cov=1e12 * cov) for x in X] for y in Y])

plt.figure(figsize=(10, 6))
plt.title("Favorite solution distribution")
plt.scatter(x_best_feasible[0], x_best_feasible[1], s=200, marker='x', label="Best feasible, f=1.34")
plt.scatter(x_post_opt[0], x_post_opt[1], s=200, marker='x', label="Post optimization, f=0.75")
plt.scatter(mean[0], mean[1], s=200, marker='x', label="Favorite, f=0.75")
plt.contour(X, Y, Z, levels=30, linestyles="dashed")
plt.plot(x, g1(x), label="g1=0")
plt.plot(2 * [np.sqrt(0.5)], [rg, -rg], label="g2=0")
plt.ylim((-0.5, rg))
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()

Z = np.array([[sum(np.square(np.array(g([x, y])))) for x in X] for y in Y])

plt.figure(figsize=(10, 6))
plt.title("Optimization landscape during post optimization")
plt.scatter(x_best_feasible[0], x_best_feasible[1], s=200, marker='x', label="Best feasible, f=1.34")
plt.scatter(x_post_opt[0], x_post_opt[1], s=200, marker='x', label="Post optimization, f=0.75")
plt.contour(X, Y, Z, levels=100, linestyles="dashed")
plt.plot(x, g1(x), label="g1=0")
plt.plot(2 * [np.sqrt(0.5)], [rg, -rg], label="g2=0")
plt.ylim((-0.5, rg))
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()

plt.show()
```
